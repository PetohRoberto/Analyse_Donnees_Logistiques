```python
import tkinter as tk
from tkinter import filedialog, messagebox
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
from tkinter import ttk
```

```python


class GraphiteInterface(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Graphite Interface")

        # Créer la gestion des onglets
        self.tabControl = ttk.Notebook(self)

        # Onglet Profilage
        self.tab1 = ttk.Frame(self.tabControl)
        self.tabControl.add(self.tab1, text='Profilage')

        # Créer la mise en page pour l'onglet Profilage
        self.left_frame = tk.Frame(self.tab1)
        self.left_frame.grid(row=0, column=0, padx=10, pady=10, sticky='nsew')

        right_frame = tk.Frame(self.tab1)
        right_frame.grid(row=0, column=1, padx=10, pady=10, sticky='nsew')

        self.load_button = tk.Button(self.left_frame, text="Charger un fichier", command=self.load_file)
        self.load_button.grid(row=0, column=0, padx=10, pady=10, sticky='e')

        # Utilisation d'une liste de labels pour afficher les colonnes
        self.column_labels = []

        self.notebook_columns = ttk.Notebook(self.left_frame)
        self.notebook_columns.grid(row=1, column=0, padx=10, pady=10, sticky='nsew')

        # Ajouter un bouton pour générer le profilage
        self.generate_button = tk.Button(self.left_frame, text="Générer le Profilage", command=self.generate_profile)
        self.generate_button.grid(row=0, column=2, padx=10, pady=10, sticky='w')

        # Onglet Analyse
        self.tab2 = ttk.Frame(self.tabControl)
        self.tabControl.add(self.tab2, text='Analyse')

        # Onglet Integration
        self.tab3 = ttk.Frame(self.tabControl)
        self.tabControl.add(self.tab3, text='Integration')

        # Pack des onglets
        self.tabControl.pack(expand=1, fill="both")

        # DataFrame actuel
        self.current_dataframe = None
        self.tab_columns = []

    def load_file(self):
        file_path = filedialog.askopenfilename(filetypes=[
            ("CSV files (UTF-8)", "*.csv"),
            ("CSV files (ISO-8859-1)", "*.csv"),
            ("Excel files", "*.xlsx;*.xls"),
            ("All files", "*.*")
        ])

        if file_path:
            try:
                # Ici, vous pouvez ajouter la logique pour traiter le fichier
                # Par exemple, lire le fichier avec pandas, etc.

                # Charger les colonnes du fichier CSV
                if file_path.lower().endswith('.csv'):
                    self.current_dataframe = pd.read_csv(file_path, encoding='utf-8')
                    self.update_column_labels()
            except Exception as e:
                # Gérer les erreurs lors du chargement du fichier
                tk.messagebox.showerror("Erreur de chargement", f"Erreur lors du chargement du fichier : {str(e)}")

    
    def update_column_labels(self):
        # Effacer les labels de colonnes actuels
        for label in self.column_labels:
            label.destroy()

        # Ajouter un label pour chaque colonne du DataFrame
        if self.current_dataframe is not None:
            sub_notebook = ttk.Notebook(self.tab1)
            sub_notebook.grid(row=0, column=0)
            for _, column in enumerate(self.current_dataframe.columns):
                tab_onglet = ttk.Button(sub_notebook)
                sub_notebook.add(tab_onglet, text=column)


    def generate_profile(self):
        # Fonction pour générer le profilage, vous pouvez ajouter votre logique ici
        tk.messagebox.showinfo("Profilage généré", "Le profilage a été généré avec succès!")

if __name__ == "__main__":
    app = GraphiteInterface()
    app.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

class VerticalTabsExample(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Vertical Tabs Example")

        # Créer le gestionnaire d'onglets
        self.tabControl = ttk.Notebook(self)

        # Onglet 1
        tab1 = ttk.Frame(self.tabControl)
        self.tabControl.add(tab1, text='Onglet 1')

        # Onglet 2
        tab2 = ttk.Frame(self.tabControl)
        self.tabControl.add(tab2, text='Onglet 2')

        # Onglet 3
        tab3 = ttk.Frame(self.tabControl)
        self.tabControl.add(tab3, text='Onglet 3')

        # Disposer les onglets verticalement
        self.tabControl.grid(row=3, column=0, sticky='ns')

        # Ajouter le contenu pour chaque onglet
        label1 = tk.Label(tab1, text='Contenu de l\'onglet 1')
        label1.pack(padx=10, pady=10)

        label2 = tk.Label(tab2, text='Contenu de l\'onglet 2')
        label2.pack(padx=10, pady=10)

        label3 = tk.Label(tab3, text='Contenu de l\'onglet 3')
        label3.pack(padx=10, pady=10)

if __name__ == "__main__":
    app = VerticalTabsExample()
    app.mainloop()
```

```python
from kivy.app import App
from kivy.uix.tabbedpanel import TabbedPanel, TabbedPanelItem
from kivy.uix.label import Label
from kivy.uix.boxlayout import BoxLayout

class VerticalTabsApp(App):
    def build(self):
        # Créer le widget TabbedPanel
        tabbed_panel = TabbedPanel(do_default_tab=False, tab_width=200, pos_hint={'center_x': 0.5, 'center_y': 0.5})

        # Onglet 1
        tab1 = TabbedPanelItem(text='Onglet 1', font_size=20)
        tab1_content = BoxLayout(orientation='vertical')
        tab1_content.add_widget(Label(text='Contenu de l\'onglet 1', font_size=16))
        tab1.content = tab1_content
        tabbed_panel.add_widget(tab1)

        # Onglet 2
        tab2 = TabbedPanelItem(text='Onglet 2', font_size=20)
        tab2_content = BoxLayout(orientation='vertical')
        tab2_content.add_widget(Label(text='Contenu de l\'onglet 2', font_size=16))
        tab2.content = tab2_content
        tabbed_panel.add_widget(tab2)

        # Onglet 3
        tab3 = TabbedPanelItem(text='Onglet 3', font_size=20)
        tab3_content = BoxLayout(orientation='vertical')
        tab3_content.add_widget(Label(text='Contenu de l\'onglet 3', font_size=16))
        tab3.content = tab3_content
        tabbed_panel.add_widget(tab3)

        return tabbed_panel

# Instancier et exécuter l'application
app = VerticalTabsApp()
app.run()
```
